```python
#visualisai total konsumsi masing masing negara
import matplotlib.pyplot as plt

energy_consumption = df.groupby("Country")["Total Energy Consumption (TWh)"].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(energy_consumption.index, energy_consumption.values, color='skyblue')

# Labels and title
plt.xlabel("Country", fontsize=12)
plt.ylabel("Total Energy Consumption (TWh)", fontsize=12)
plt.title("Top 10 Total Energy Consumption by Country", fontsize=14)
plt.xticks(rotation=90)

plt.show()
```

```python
# We'll check in which year did we have to most total consumption
# untuk mengecek berapa total konsumsi setiap tahunnya

df.groupby("Year")["Total Energy Consumption (TWh)"].sum()
```

```python
# visualisasi total enersi consumption per tahun
import matplotlib.pyplot as plt

yearly_consumption = df.groupby("Year")["Total Energy Consumption (TWh)"].sum()

x = yearly_consumption.index
y = yearly_consumption.values

plt.figure(figsize=(12, 6))
plt.bar(x, y)

plt.xticks(ticks=x, labels=x, rotation=90, fontsize=10)

plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Energy Consumption (TWh)", fontsize=12)
plt.title("Yearly Energy Consumption", fontsize=14)

plt.show()
```

```python
# visualisai total harga per capita
import matplotlib.pyplot as plt

PCE = df.groupby("Country")["Per Capita Energy Use (kWh)"].mean().sort_values(ascending=True)

x = PCE.index
y = PCE.values

plt.figure(figsize=(12,6))
bars = plt.bar(x, y, color="blue")

plt.xticks(rotation=90, fontsize=10)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Per Capita Energy Use (kWh)", fontsize=12)
plt.title("Per Capita Energy Use Across Countries", fontsize=14)

plt.bar_label(bars, fmt="%.2f", fontsize=10, padding=3, color="black")

plt.show()
```

```python
#pemakaian energi pada masing2 negara
countries = df["Country"].unique()

for country in countries:
    country_data = df[df["Country"] == country]

    industrial_use = country_data["Industrial Energy Use (%)"].mean()
    household_use = country_data["Household Energy Use (%)"].mean()

    labels = ["Industrial", "Household"]
    values = [industrial_use, household_use]

    plt.figure(figsize=(3,3))
    plt.pie(values, labels=labels, autopct="%.1f%%", colors=["pink", "red"])
    plt.title(f"Energy Use Distribution in {country}")

    plt.show()
```

```python
# 1. Which countries have the highest share of renewable energy?

RE = df.groupby("Country")["Renewable Energy Share (%)"].mean().sort_values(ascending=True)

x = RE.index
y = RE.values
plt.figure(figsize=(12,6))
bars = plt.bar(x, y)

plt.xlabel("Country")
plt.xticks(rotation=45)
plt.ylabel("Renewable Energy Share (%)")

plt.bar_label(bars, fmt="%.2f", fontsize=10, padding=3, color="black")
plt.show()
```

```python
df.head()
```

```python
# Has the renewable energy share increased over time?

REOT = df.groupby("Year")["Renewable Energy Share (%)"].mean()
x = REOT.index
y = REOT.values

plt.bar(x, y)
```

```python
import seaborn as sns
data_grouped = df.groupby("Country")[["Renewable Energy Share (%)", "Carbon Emissions (Million Tons)"]].mean().reset_index()

data_melted = data_grouped.melt(id_vars="Country", var_name="Energy Type", value_name="Percentage")

plt.figure(figsize=(12, 6))
sns.barplot(data=data_melted, x="Country", y="Percentage", hue="Energy Type")

plt.xticks(rotation=45)
plt.title("Energy Share by Country")
plt.tight_layout()
plt.show()
```

```python
#total carbon emisi
TCE = df.groupby("Country")["Carbon Emissions (Million Tons)"].sum().sort_values(ascending=True)

x = TCE.index
y = TCE.values

plt.figure(figsize=(12,6))
bars = plt.bar(x, y)

plt.xlabel("Country")
plt.ylabel("Carbon Emissions (Million Tons)")

plt.show()
```

```python
# 1. How do energy prices vary by country?
EPI = df.groupby("Country")['Energy Price Index (USD/kWh)'].mean().sort_values(ascending=True)

x = EPI.index
y = EPI.values

plt.scatter(x, y)
plt.xticks(rotation = 45)
plt.xlabel("Countries")
plt.ylabel("Energy Price Index (USD/kWh)")
```
